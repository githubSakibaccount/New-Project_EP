# Exploratory Data Analysis

```python
import matplotlib.pyplot as plt

# Get value counts for stroke column
counts = Earthquack_clean['tsunami'].value_counts()

# Set custom labels for pie chart
labels = ['No Tsunami', 'Tsunami']

# Create pie chart with custom colors and labels
colors = ['lightblue', 'Blue']
plt.pie(counts, colors=colors, labels=labels, autopct='%1.1f%%', startangle=30)

# Add plot title
plt.title("Value count distribution of tsunami attribute")

# Show plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(4,4))
plt.title("Number of tsunami occurs in Earthquack")
sns.countplot(data=Earthquack_clean,x='tsunami',hue='tsunami')
plt.show()
```
